### 2. التحفيز: تجربة رمي العملة

**الموقف**  
عندنا عملة. مش عارفين هي fair (يعني تجيب heads قد ما تجيب tails) ولا لأ. هنرميها كذا مرة ونحاول نقدّر θ (theta) - وهي نسبة الـ heads.

**التجربة الأولى:** ١٠ رميات ← ٥ heads، ٥ tails ← θ = ٥/(٥+٥) = ٠.٥  
**التجربة الثانية:** ١٠٠ رمية ← ٥٠ heads، ٥٠ tails ← θ = ٠.٥  
**التجربة التالتة:** ١٠٠٠ رمية ← ٥٠٠ heads، ٥٠٠ tails ← θ = ٠.٥

**المشكلة مع Point Estimates**  
لو حد طلب منك تبلغ عن θ في تقرير، أي تجربة منهم هتكتبها وأنت مطمن أكتر؟

- التلات تجارب كلها أعطت θ = ٠.٥  
- بس حاسس إن التجربة بتاعة ١٠٠٠ رمية هي الأوثق  
- الإحساس بالثقة ده مش قادر تحط له أرقام (مش quantified)  
- المعالجة الرياضية هنا incomplete

**الإستنتاج:** إحنا محتاجين أكتر من point estimate. محتاجين probability distribution حوالين التقدير.

### 3. من Point Estimates لـ Probability Distributions

**تصوير الثقة بالـ Distributions**  
لما نتعامل مع θ كـ probability distribution بدل ما هي نقطة واحدة:

- Distribution A (عريض/مسطّح): ثقة قليلة - قيم كتيرة لـ θ ممكن تكون بنفس الاحتمال  
- Distribution B (ضيق/حاد): ثقة أعلى - θ مركز حوالين ٠.٥  
- Distribution C (أضيق): ثقة عالية جداً - θ تقريباً أكيد ٠.٥

**حالة Uniform Distribution**  
الـ uniform distribution (مسطّح تماماً) بيعبر عن أقصى uncertainty:  
- كل القيم بين ٠ و ١ ليها نفس الاحتمال  
- مينفعش تاخد قرارات واثقة بناءً على كده  
- ده اسمه "non-informative" distribution

**تطبيق خطير:** في قرارات مصيرية (زي جرعة دواء ليها آثار جانبية)، وجود flat distribution بيخلّي إتخاذ القرار مستحيل تقريباً، لأن كل النتائج المحتملة ليها probabilities متقاربة.

**الخلاصة**  
"إحنا مش بنعمل modeling في الماشين ليرنين عشان نحصل على point estimates. لا، إحنا بنعمل modeling عشان نحصل على probability distributions."

### 4. Linear Regression: الإفتراض الخفي

**الطريقة التقليدية**  
معادلة الانحدار الخطي: y = a + bx  
الهدف: إيجاد المعاملات a (intercept) و b (slope) اللي يقللوا loss function (غالباً MSE)

**الخطوات:**  
١. عندنا نقاط بيانات (x, y)  
٢. نحسب الخطأ: y - ŷ (حيث ŷ = a + bx)  
٣. نربّع الأخطاء  
٤. نقلّل مجموع الأخطاء المربعة عن طريق تعديل a و b  
٥. الناتج: قيم a و b

**مشكلة Point Estimate بتكرر**  
لما نستخدم الـ model المدرّب عشان نتوقع y لقيمة x جديدة:  
- بنجيب قيمة واحدة (point estimate)  
- لكن بص على النقاط الأصلية حوالين خط الانحدار  
- هما شبه "سحابة" (cloud) حوالين الخط  
- عرض السحابة ده بيعبّر عن uncertainty

**مثال:** لو x = ١٠، الـ prediction هي y = ٣٢. بس هل ممكن y تكون ٣٢.٣؟ ٣١.٨؟ أيوه - لأن النقاط الأصلية منتشرة حوالين الخط.

**ليه ده مهم لإتخاذ القرارات**  
لو الـ prediction ده هيستخدم في قرارات عملية:  
- لو السحابة ضيقة: ثقة عالية في التوقع  
- لو السحابة واسعة: ثقة قليلة - التوقع ممكن يبعد كتير عن الواقع

**أمثلة على التأثير:**  
- إدارة المخزون: لو خزنت بضاعة زيادة (فلوس راكدة) أو خزنت أقل (عملاء زعلانين)  
- قرارات مالية: مخاطرة الإستثمارات  
- تطبيقات طبية: نتائج علاج المرضى

**الحتة الناقصة**  
في تعليم الـ linear regression التقليدي:  
- إحنا بنركّز على إني أجيب a و b صح  
- نادراً ما بنتكلم عن عرض السحابة (الـ uncertainty)  
- المعاملات ممكن تكون مضبوطة، بس إحنا فوتّنا quantification of uncertainty

**الواقع:** عند كل نقطة x، في Normal distribution بيوصف احتمالية قيم y المختلفة. اللي ناقص إيه؟ σ (sigma) - الـ standard deviation اللي بيوصف الإنتشار.

### 5. Heteroscedasticity: لما الـ Noise مش ثابت

**إفتراض ثبات الـ Noise**  
الـ linear regression التقليدي بيفترض إن الـ noise (σ) ثابتة لكل قيم x:  
- الإنتشار حوالين الخط هو هو في كل الأماكن  
- الإفتراض ده هو اللي يخلّينا نقدر نحصل على closed-form solution (normal equation: XᵀX)

**لما الإفتراضات تنكسر**  
سيناريو حقيقي: إيه لو الـ noise بتتغير مع x؟  
- في بعض قيم x، التوقعات دقيقة (uncertainty قليلة)  
- وفي قيم x تانية، التوقعات واسعة (uncertainty عالية)  
- الـ noise بقت function of x (heteroscedasticity)

**المشكلة في التصور**  
لو دخلت بيانات فيها noise مش موحّدة لخوارزمية linear regression عادية:  
- هيحسب residuals وياخد متوسطهم  
- هيبالغ في الـ noise في مناطق (overestimation)  
- هيقلل من الـ noise في مناطق تانية (underestimation)  
- ورغم كده، a و b ممكن يطلعوا صح رياضياً!

**التسليم الخطير**  
تخيل إنك سلمت معادلة الـ regression (a و b) لفريق في البنك أو في المستشفى:  
- هما هيستخدموها في كل الأماكن بنفس الثقة  
- في المناطق اللي الـ noise فيها قليل، التوقعات تمام  
- في المناطق اللي الـ noise فيها عالي، التوقعات فاشلة  
- إنت ادّيتهم نص الحقيقة (المتوسط بس، من غير الإنتشار)

**المصطلح:** لما الـ noise يختلف مع x، ده اسمه heteroscedasticity - مشكلة مشهورة جداً في الـ data science.

**السؤال الكبير**  
إزاي نجيب معادلة الـ noise؟ إزاي نعمل model للـ σ كـ function of x؟  
لو إفترضنا شكل معيّن للدالة (زي quadratic مثلًا)، وطلع الإفتراض غلط، النتائج هتبقى كارثية. أحياناً الـ noise بيجي من مصادر معقّدة (البيئة المحيطة، تصرفات المنافسين، ديناميكيات السوق) ومش سهل نمذجتها.  
هنا بنحتاج آليات تقدر "تلتقط" الـ noise من غير ما نفترض شكل الدالة.

### 6. خطر Closed-Form Solutions

**تحذير مهم جداً**  
"ما تقبلش أي closed-form solution إلا لما تحقق الإفتراضات اللي هو قائم عليها."  

ليه؟ الـ closed-form solutions (زي normal equation بتاعة linear regression) بيخبّوا جواهم "وش قبيح". عشان نوصل للحل المغلق ده، علماء الرياضيات عملوا افتراضات. لو الإفتراضات دي مش مطابقة لحالتك الحقيقية، النتايج ممكن تكون منحرفة بشكل كبير.

**القاعدة:** دايماً حقّق ورا أي analytical solution قبل ما تطبقه على مشكلتك.

### 7. Maximum Likelihood Estimation (MLE)

**التحول الجذري**  
إحنا محتاجين ننتقل من optimizing لـ point estimates لـ optimizing لـ probability distributions. الـ loss function اللي بتكلم عن قيم y مش هتنفع. إحنا محتاجين loss function بتيج على بالها probability.

**أهلاً بـ Maximum Likelihood Estimation (MLE)**

**بناء الـ Model بطريقة Probabilistic**  
عندنا:  
- مجموعة مدخلات (x)  
- مجموعة مخرجات (y)  

بس دلوقتي عايزين نعمل model لـ P(y) مش بس y.  
فاكر إزاي عند كل x في Normal distribution واقفة راسية؟ لكل نقطة بيانات، في probability إننا نشوف قيمة y دي مع x دي.

**الفكرة الأساسية**  
إحنا عايزين الـ model بتاعنا يكون قادر يعبّر عن قيم y المختلفة. بكلمات تانية، عايزين probability of the data تكون أكبر ما يمكن.  
لو قدرنا نكبر الـ joint probability لكل قيم y اللي شفناها (y₁, y₂, y₃, ..., yₙ)، يبقى كده لقينا parameters اللي تخلّي البيانات دي "most likely".  

ده هو Maximum Likelihood Estimation - نفس السحر اللي شغال في كل حاجة من linear regression لـ large language models.

**إفتراض الـ Independence**  
بنفترض عادة إن قيم y مستقلة عن بعضها:  
- مفيش dependency بين قياس والتاني  
- ده إفتراض منطقي في حالات كتير (زي قياس أطوال ناس عشوائيين)  
- إستثناء: تحليل السلاسل الزمنية (time series) - هنا القيم بتعتمد على اللي قبلها  

مع الـ independence، الـ joint probability بقت حاصل ضرب الـ probabilities الفردية:  
P(y₁, y₂, ..., yₙ) = P(y₁) × P(y₂) × ... × P(yₙ)  

كل P(yᵢ) جواها الـ term (a + bxᵢ) (كـ mean للتوزيع). فبقى عندنا معادلة فيها a و b نقدر نكبر قيمتها.

### 8. الفرق بين Likelihood و Probability

**التمييز الجوهري**

**Probability Distribution:**  
- إنت عارف الـ parameters (μ, σ)  
- عايز تعرف P(x) لقيم x مختلفة  
- x على المحور الأفقي  
- Probability على المحور الرأسي  

**Likelihood:**  
- إنت عارف البيانات (قيم x اللي شفتها)  
- عايز تعرف الـ parameters (μ, σ) اللي تخلّي البيانات دي "most probable"  
- الـ parameters (μ, σ) على المحور الأفقي  
- عند كل combination من parameters، بتحسب حاصل ضرب probabilities كل نقاط البيانات

**تصوّر الـ Likelihood**  
تخيّل عندك histogram للبيانات (أزرق). بتحاول تركّب Gaussian distribution (خط أحمر) عليه:  
- حرّك الـ mean (μ) شمال/يمين  
- عدّل الـ standard deviation (σ)  
- عند كل combination، شوف الخط الأحمر قد إيه مطابق للـ histogram الأزرق  
- الـ combination اللي يعطي أفضل مطابقة هو Maximum Likelihood Estimate  

**مثال:** لو البيانات مركزها عند ١٠ و σ=٢، فـ setting μ=١٠ و σ=٢ هيعطي أعلى قيمة likelihood. أي deviation (μ=١١ مثلًا) هيقلل الـ likelihood.

**ليه الـ Likelihood هي "Probability مقلوبة"**  
في probability: Parameters ← Data  
في likelihood: Data ← Parameters  
القلب ده صعب إستيعابه شوية، لكنه قوي رياضياً.

### 9. خدعة Log-Likelihood

**مشكلة الـ Numerical Stability**  
الـ probabilities أرقام أقل من ١. لما تضرب probabilities كتير مع بعض (آلاف البيانات)، الناتج بيبقى صغير جداً - يقترب من الصفر.  
المشكلة: الكمبيوتر مش بيعرف يتعامل بدقة مع الأرقام الصغيرة جداً دي (حدود floating-point arithmetic).

**الحل: اللوغاريتمات**  
الـ log بتحوّل الضرب لجمع:  
log(A × B) = log(A) + log(B)  

بدل ما نكبر الـ likelihood (حاصل ضرب probabilities)، بنكبر الـ log-likelihood (مجموع log probabilities).  

**خاصية مهمة:** الـ maximum of a function بيحصل عند نفس النقطة زي maximum of its log. فـ maximizing log-likelihood بيدينا نفس الـ parameters اللي كنا هناخدها من maximizing likelihood.  
دي transformation رتيبة (monotonic) - آمنة رياضياً ومستقرة عددياً.

### 10. من متغير واحد للأبعاد العالية (High Dimensions)

**التحدي**  
لغاية دلوقتي كنا بنتكلم عن random variable واحد. لكن المشاكل الحقيقية فيها آلاف بل ملايين المتغيرات:  
- الصور: آلاف الـ pixels  
- النصوص: آلاف الـ tokens/words  
- بيانات الحساسات: قياسات متعددة في نفس الوقت  

شكل الـ probability distribution في ١٠,٠٠٠ بعد (dimension) شكله إيه؟ مفيش كتالوج أو توزيعات جاهزة للمساحات عالية الأبعاد دي.

**الطريق للأمام**  
إحنا محتاجين ننتقل من:  
- عندي probability distributions لمتغيرات عشوائية فردية  
- لبناء probability distributions لأنظمة عالية الأبعاد فيها ملايين المتغيرات العشوائية  

**الأداة:** Probabilistic Graphical Models (PGMs)

### 11. Probabilistic Graphical Models (PGMs)

**الفكرة الأساسية**  
خد بياناتك واكسرها لمجموعة من الـ random variables. حدّد العلاقات بين المتغيرات دي. مثّل العلاقات دي في شكل graph:  
- Nodes: المتغيرات العشوائية  
- Edges: الإعتمادات (dependencies) بين المتغيرات

**من المعادلات للـ Graphs**  
عندنا نظام معادلات:  
```
x₁ ~ Normal(μ=5z, σ=10)  
x₂ ~ Normal(μ=0.5x₁, σ=1)  
...
```  
كل معادلة بتمثل probability distribution لمتغير عشوائي، ويمكن يكون dependent على متغيرات تانية.  

النظام ده ممكن تمثيله بالـ graph:  
- z → x₁ (edge من z لـ x₁)  
- x₁ → x₂ (edge من x₁ لـ x₂)

**قوة التمثيل بالـ Graph**  
الـ graph بيوضّح الـ dependency structure بصرياً. بيورينا:  
- إيه المتغيرات اللي بتعتمد على إيه  
- تدفق التأثير خلال النظام  
- علاقات conditional independence

**بناء Generative Model**  
لما يكون عندك:  
١. Random variables محددة  
٢. Dependency structure مرسوم (الـ graph)  
٣. Functional form إختارتها لكل متغير (وفيها parameters هنتعلمها)  
يبقى عندك Generative Model - نموذج فاهم الـ probability distribution لكل المتغيرات مع بعض.

**عملية الـ Learning**  
الـ learning = إيجاد الـ parameters المجهولة في الـ functional forms  
إزاي؟ عن طريق maximizing الـ likelihood (joint probability) لكل البيانات اللي شفناها.  

لما الـ parameters تتعلم، تقدر:  
- تولّد بيانات جديدة (sampling)  
- تعمل predictions (conditioning)

---

### 12. Ancestral Sampling و Querying

**Sampling من الـ Graph**  
عندك PGM، إزاي تولّد عينات جديدة؟  

**Ancestral Sampling:** إبدأ من الـ nodes اللي معندهاش parents (root nodes)، خذ samples منهم، وبعدين إمشي لأولادهم، وإستخدم قيم الـ parents اللي أخذتها عشان تحدّد distributions بتاعة الأولاد.

**مثال مع variables x₁ لحد x₅:**  
١. خذ sample من x₁ (مفيش parents)  
٢. خذ sample من x₂ conditioned على x₁ (في edge)  
٣. خذ sample من x₃ conditioned على x₁ و x₂  
٤. خذ sample من x₄ conditioned على parents بتوعه  
٥. خذ sample من x₅ conditioned على x₂ و x₄  

مينفعش تاخد sample من node إلا بعد ما تكون كل parents بتوعها اتعملها sample.

**Querying: آلية الـ Prediction**  
لما يكون عندك generative model مدرب، تقدّم تسأل أسئلة:  
مثال: "إيه توزيع x₁ لو x₂ = قيمة كذا؟"  

**الطريقة:**  
١. طبّق الـ condition (filter) على x₂  
٢. بص على الـ samples الناتجة لـ x₁ اللي تحقق الشرط ده  
٣. الـ histogram للعينات المفلترة دي هو conditional distribution P(x₁ | x₂)

**نقطة جوهرية:** في الـ generative modeling، إنت مش بتحدّد مسبقاً إيه الـ predictors وإيه الـ responses. إنت بتنبني الـ joint distribution لكل حاجة، وبعدين بتعمل condition على أي variables شايفها. الـ variables اللي متبقية هما predictions بتاعتك.

**العلاقة بـ ChatGPT**  
ده بالظبط إشتغال large language models:  
- هما بيعملوا probabilistic model على كل الـ sequences الممكنة في اللغة الإنجليزية  
- لما تكتب prompt، إنت بتعمل condition على الـ sequence ده  
- الـ next token prediction هو الـ response conditioned على prompt بتاعك  

إنت مكنتش محتاج تقول للموديل "دول predictors ودول responses" - الـ conditioning هو اللي قام بالشغل.

---

### 13. قاعدة Chain Rule و Conditional Independence

**قاعدة Chain Rule العامة**  
لأي مجموعة random variables:  
P(x₁, x₂, ..., xₙ) = P(x₁) × P(x₂|x₁) × P(x₃|x₁,x₂) × ... × P(xₙ|x₁,...,xₙ₋₁)  

ده دايماً صح، بغض النظر عن الـ dependency structure.

**تطبيق هيكل الـ Graph**  
لما نطبّق الـ dependency structure الحقيقية من الـ graph بتاعنا، قاعدة الـ chain rule بتتختصر:  
- لو variable معيّن مستقل عن بعض الـ ancestors بمجرد ما تعرف parents بتوعه، الحدود دي بتشال  
- الصيغة العامة بقت حاصل ضرب conditional probabilities بحيث كل variable مأخوذ conditioned على parents بتوعه فقط  

P(x₁, x₂, ..., xₙ) = Π P(xᵢ | parents(xᵢ))

**مثال تطبيقي**  
لـ graph فيه ٨ variables وليهم dependencies معينة:  
- x₁ معندوش parents  
- x₂ dependent على x₁  
- x₃ dependent على x₁ بس  
- x₄ dependent على x₂ و x₃  
- x₅ dependent على x₄  
- ... إلخ  

الـ joint probability بقت:  
P(x₁) × P(x₂|x₁) × P(x₃|x₁) × P(x₄|x₂,x₃) × P(x₅|x₄) × ...

**الإختراق**  
التحليل ده هو جوهر probabilistic graphical modeling:  
"الـ joint probability لنظام من random variables هو حاصل ضرب probability of each random variable conditioned on its parents."  

من هنا:  
١. بنبني likelihood function (حاصل ضرب الحدود دي)  
٢. ندخل البيانات اللي شفناها  
٣. نعمل optimization عشان نلاقي الـ parameters  
٤. يبقى عندنا generative model كامل

---

### 14. بناء Joint Probability Distribution

(مدموج مع القسم ١٣ كجزء أساسي من الـ PGM)  
الـ joint probability distribution تبنى بالضبط من خلال الـ Chain Rule المبسطة حسب هيكل الـ graph، وده اللي بنستخدمه في maximizing the likelihood أثناء التعلم.

---

### 15. Neural Networks كـ Function Approximators

**مشكلة Functional Form**  
في PGM بتاعنا، محتاجين نحدد functional form لكل dependency:  
- P(x₂|x₁) خطي؟ تربيعي؟ حاجة تانية؟  
- في مسائل بسيطة، خبراء المجال ممكن يعرفوا (زي الفيزيائيين أو المحللين الماليين)  
- في مسائل معقّدة، إحنا مش عارفين الـ functional form الحقيقية  

**الـ Universal Function Approximator**  
الـ neural networks تقدر تقرّب أي function لو أعطيتها parameters كافية. ده يخلّيها مثالية لتمثيل dependencies المجهولة جوا PGM.  

**الطريقة الهجينة:** إستخدم neural network components عشان تمثّل أجزاء من likelihood function اللي إحنا مش عارفين شكلها.

---

### 16. Case Study: نمذجة Unknown Noise

**إعداد المشكلة**  
نولّد بيانات فيها heteroscedasticity مقصودة:  
- الـ noise بتزيد خطياً مع x (بس إحنا بنتجاهل إن إحنا عارفين ده)  
- عايزين نبنى model يمسك الـ mean ويغيّر uncertainty حسب x

**الصياغة Probabilistic**  
لكل نقطة بيانات (x, y):  
P(y|x) = (١/√(٢πσ(x))) × exp(-(y - μ(x))²/(٢σ(x)²))  

حيث:  
- μ(x) = a + bx (mean خطي)  
- σ(x) = neural_network(x) (function مجهولة هنتعلمها)

**دالة Likelihood**  
الـ likelihood (joint probability لكل y's) هو حاصل ضرب الحدود دي لكل نقاط البيانات. وإستخدام log-likelihood عشان الـ numerical stability:  
log L = Σ log P(yᵢ|xᵢ)

**التطبيق في PyTorch**  
١. تعريف الـ parameters: a, b, و weights بتاعة neural network  
٢. Forward pass:  
   - حساب μ = a + b*x  
   - حساب σ = neural_net(x)  
   - التأكد إن σ positive (إستخدم softplus أو exp)  
٣. حساب negative log-likelihood loss  
٤. Backpropagation وتحديث كل الـ parameters

**مقارنة النتائج**  

**Standard Linear Regression (الخط البنفسجي):**  
- بيدي mean line مضبوط  
- بيدي confidence intervals ثابتة (نفس العرض في كل حتة)  
- بيعطي overestimation للـ uncertainty في مناطق الـ noise القليل  
- بيعطي underestimation للـ uncertainty في مناطق الـ noise العالي  

**Probabilistic Neural Approach (الخط البرتقالي + المنطقة المظللة):**  
- بيدي mean line مضبوط  
- بيدي confidence intervals متغيرة بتناسب تركيب الـ noise الحقيقي  
- intervals ضيقة في مناطق الـ noise القليل  
- intervals واسعة في مناطق الـ noise العالي  
- uncertainty quantification دقيق

**التحول الجذري**  
"لو إنت فهمت المثال ده - إستخدام neural network عشان تمثّل جزء من likelihood function في probabilistic graphical model - ده paradigm shift في حياتك المهنية. هتقدر تقرأ أبحاث كتير جداً بتستخدم neural components عشان تمثّل dependencies مجهولة في PGMs."

---

### 17. خلاصة عملية وإتخاذ القرارات

**إطار القرار**

**الخطوة ١: Exploratory Data Analysis**  
- بص على بياناتك وشوف forms  
- دوّر على patterns في residuals/noise  
- تأكد من الإفتراضات (ثبات variance، independence، إلخ)

**الخطوة ٢: قياس الـ Criticality**  
- قد إيه مهم تعرّف الـ uncertainty في المشكلة دي؟  
- هل القرارات هتتغير لو التوقعات بعدت عن الحقيقة؟  
- ده تطبيق high-stakes (رعاية صحية، بنوك، سلامة)؟

**الخطوة ٣: إختر الطريقة**

<div dir="ltr">

| الموقف                              | الطريقة المناسبة                          |
|-------------------------------------|---------------------------------------------|
| الإفتراضات صحيحة، stakes قليلة     | Classical ML (سريع، closed-form)           |
| الإفتراضات صحيحة، stakes عالية     | Classical ML + uncertainty quantification   |
| الإفتراضات مش صحيحة، stakes قليلة  | فكر إذا كان الموضوع فارق                   |
| الإفتراضات مش صحيحة، stakes عالية  | Probabilistic ML (PGMs, MLE, neural components) |

</div>

**مبادئ أساسية**  
١. ما تثقش في point estimates على عميان - هما بيخبّوا uncertainty  
٢. حقّق ورا الإفتراضات اللي أي closed-form solution مبني عليها  
٣. طابق الطريقة مع المشكلة - probabilistic ML قوي بس مكلف حسابياً  
٤. إعمل validation كتير - الـ test set لازم يثبت إن الـ model بيعمم كويس  
٥. خبرة المجال مهمة - إختيار distributions مناسبة للمتغيرات محتاج فهم لطبيعة البيانات

**عقلية الـ Industry**  
في الـ industry، شغلك هو:  
- تحل مشاكل مالهاش حلول واضحة  
- تكسّر المشاكل المعقّدة لمكونات  
- تقدم إجابات اللي بيتخذوا القرارات يقدر يثقوا فيها  
- تنقل الـ uncertainty بشكل مناسب  

إنت مش بالضرورة بتعمل research جديد - إنت بتستخدم الأدوات الموجودة بذكاء، مع وعي بإفتراضاتهم وحدودهم.

---

### 18. أبرز أسئلة Q&A

**س١:** إزاي أعرف إن إفتراضاتي عن الـ dependencies غلط؟  
**ج:** نفس طريقة validation في أي ML model - من خلال validation و test sets. لو الـ model أداؤه وحش على test data، فإفتراضاتك (ومنها dependency structure) ممكن تكون غلط. في حالات متقدمة، ممكن تعمل model selection - تعامل الـ model نفسه كـ parameter وتتعلّم أي structure هو الأحسن (بس ده مكلف حسابياً).

**س٢:** هل الـ reinforcement learning ينفع لإختيار الـ model؟  
**ج:** تكلفة تدريب RL agent عشان يعمل model selection غالباً هتزيد عن تكلفة تدريب نماذج متعددة وإختبارهم. الـ validation التقليدي عادةً أعملي.

**س٣:** إيه العمل لو الـ samples dependent على بعضها (زي time series أو أنظمة مالية)؟  
**ج:** دي اسمها dynamical systems. في السلاسل الزمنية، كل قيمة بتعتمد على اللي قبلها. الإعتماد ده بيبقى جزء من هيكل الـ model (زي autoregressive models). نفس إطار PGM بيشتغل - بس بتضيف الـ dependencies الزمنية كـ edges في الـ graph بتاعك.

**س٤:** ليه مش بحسب μ و σ مباشرة من البيانات بمعادلات بسيطة؟  
**ج:** الطريقة دي بتشتغل لما تكون متأكد إن التوزيع Gaussian وعندك variable واحد. لكن لما يكون الـ likelihood حاصل ضرب distributions معقّدة كتير (زي Bernoulli، Gamma، وغيره)، مفيش معادلة بسيطة. Maximum likelihood estimation عن طريق optimization هو اللي بيعالج الحالات المعقّدة دي.

**س٥:** إيه الفرق بين الـ industry والـ academia في التدقيق على الإفتراضات؟  
**ج:** في ناس في الـ industry بيهملوا الإفتراضات تماماً، ويثقوا فقط في validation metrics. ده خطير لأن:  
- بيانات test النهاردة ممكن متبقاش ممثلة لبيانات بكرة (model drift)  
- توزيعات البيانات المختلفة ممكن تكسر الإفتراضات بصمت  
- ممكن تجربتك تنجح مع dataset معيّن وتفشل مع غيره  
**المنهج المتوازن:** إفهم الإفتراضات، وإعمل validation كامل، وراقب أداء الـ model مع الوقت.

**س٦:** إزاي المبتدئ يخطط لطريقته في التعلم؟  
**ج:** الموضوع شخصي جداً. إستخدم ChatGPT/Gemini عشان تعمل personalized plans:  
- قوله إيه اللي إنت عارفه  
- حدد إيه إنت عايز توصله  
- أطلب منه resources مرتبة في sequence  
- الـ AI ممكن يعمل خطط أحسن من أي نصيحة عامة ١٠ مرات

**س٧:** هل لازم أكون متمكن من كل الرياضيات قبل ما أقرأ research papers؟  
**ج:** لا. وأثناء قراية الأبحاث، حتماً هتقابلك مفاهيم رياضية جديدة. ده طبيعي. قف، إتعلم المفهوم ده بالذات (يمكن تاخد أسبوع أو إثنين)، بعدين إكمل. كلنا بنعمل كده - حتى الباحثين الخبرة.

**س٨:** هل نقدر نولّد أنواع بيانات مختلفة (صور من نص، إلخ)؟  
**ج:** أيوه. الـ random variables بتاعتك هي اللي بتحدد أنواع البيانات. لو عندك pixel random variables و token random variables وبينهم dependencies مناسبة، الـ model بتاعك يبقى multi-modal. إنت بتحدد المشكلة، والـ framework بيهندلها - بشرط يكون عندك قدرة حسابية كافية.

---

### 19. مصادر وكتب مقترحة

**كتب**  
١. **"Bayesian Reasoning and Machine Learning"** David Barber  
   - مقدمة لطيفة في التفكير probabilistic في ML  
   - نقطة بداية كويسة قبل التعمق  

٢. **"Deep Learning Foundations and Concepts"** Christopher Bishop (٢٠٢٣)  
   - يربط الـ deep learning بـ probabilistic programming  
   - تحميل مجاني من Springer  
   - أقل كثافة من كتابه القديم (٢٠٠٩)  

٣. **"Probabilistic Machine Learning"** Kevin Murphy  
   - جزئين: Introduction و Advanced Topics  
   - حوالي ٢٠٠٠ صفحة  
   - مرجع شامل  
   - لو فهمت الأفكار الأساسية من الجلسة دي، إنت جاهز للكتب دي

**إتجاهات بحثية**  
- Structure Learning: Algorithms بتحاول تتعلم الـ dependency structure من البيانات (لسه صعبة - النتائج متواضعة غالباً)  
- Dynamic Systems: معالجة الـ dependencies الزمنية  
- Neural PGMs: إستخدام neural networks عشان تمثّل functional forms مجهولة جوا probabilistic graphical models  
- Multi-modal Models: دمج أنواع بيانات مختلفة (صور، نص، صوت) في إطار probabilistic واحد



</div>
